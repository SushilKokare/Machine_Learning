```python
# Random Data Distribution
```

```python
import numpy as np
```

```python
x=np.random.normal(5,1,1000) # mean=5 , sd=1 , size=1000
y=np.random.normal(10,2,1000)
```

```python
plt.scatter(x,y)
plt.show()
```
